### Connecting Colab to GitHub

1.  **Save a copy to GitHub**: Go to `File` > `Save a copy to GitHub`. This will prompt you to authorize Colab to access your GitHub repositories if you haven't already.
2.  **Choose Repository and Filename**: Select the desired GitHub repository and provide a filename for your notebook. You can also create a new repository directly from this dialog.
3.  **Connect to GitHub**: Once the initial save is done, you'll see a 'GitHub' icon in the Colab toolbar. Clicking this will allow you to:
    *   **Open a notebook from GitHub**: Browse your repositories to open an existing `.ipynb` file.
    *   **Save a copy to GitHub**: Save the current notebook as a new file in a GitHub repository.
    *   **Save a copy in GitHub as a new version**: If you opened the notebook from GitHub, this option will be available to commit changes back to the original file in the repository.

Alternatively, you can go to `File` > `Open notebook` and select the `GitHub` tab to browse and open notebooks directly from your repositories.
